##Bayes Theorem

$ P(\text{A} \mid \text{B}) = \frac{P(\text{B} \mid \text{A}) \cdot P(\text{A})}{P(\text{B})} $

P(A/B): Posterior probability of class A given predictor B

P(B/A): Likelihood of predictor B given class A

P(B/A) = ||B| + |A|| / |A|

P(A): Prior probability of class A

P(B): Prior probability of predictor B

## Problem 1:

Joan says she is itchy. There is a test for Allergy to Cats, but this test is not always
right:

- For people that really do have the allergy, the test says "Yes" 80% of the time
- For people that do not have the allergy, the test says "Yes" 10% of the time

If 1% of the population has the allergy, and Joan’s test says "Yes", what are the chances that Joan really has the allergy?


$ P(\text{Yes} \mid \text{Allergy}) = \frac{P(\text{Allergy} \mid \text{Yes}) \cdot P(\text{Yes})}{P(\text{Allergy})} $

```python
#Given values
p_A = 0.01 #P(A): has allergy
p_notA = 1 - p_A #P(not A)

p_pos_given_A = 0.80 #P(+ | A)
p_pos_given_notA = 0.10 #P(+ | not A)

print(f'''
  {p_A}
  {p_notA}

  {p_pos_given_A}
  {p_pos_given_notA}
  ''')
```

```python
#Step 1: Compute total probability of a positive test
p_pos = (p_pos_given_A * p_A) + (p_pos_given_notA * p_notA)
```

```python
#Step 2: apply Bayes' Theorem
p_A_given_pos = (p_pos_given_A * p_A) / p_pos

print(p_A_given_pos)
```

## Problem 2:

Beth is planning a picnic today, but the morning is cloudy.

- 50% of all rainy days start off cloudy
- Cloudy mornings are common (40% of all days start cloudy)
- This is usually a dry month (10% of days tend to be rainy)

What is the chance of rain during the day when it starts of cloudy?

```python
#Given values
p_R = 0.10 #P(R) 10% of days tend to be rainy
p_C_given_R = 0.50 #P(C | R) 50% of all rainy days start off cloudy
p_C = 0.40 #P(C) 40% of all days start cloudy
```

```python
p_R_given_C = (p_C_given_R * p_R) / p_C

print(p_R_given_C)
```

## Problem 3:

An artist competition has entries from three painters: Pam, Paul and Peter.

- Pam put in 15 paintings, 4% of her works have won First Prize.
- Paul put in 5 paintings, 5% of his works have won First Prize.
- Peter put in 10 paintings, 3% of his works have won First Prize.

What is the probability that Pam will win First Prize?

```python
#Prior probabilities (based on number of paintings)
p_pam = 15/30
p_paul = 5/30
p_peter = 10/30
```

```python
#Likelihoods (Win rates)
p_win_given_pam = 0.04
p_win_given_paul = 0.05
p_win_given_peter = 0.03
```

```python
#Total probability of a winning painting
p_win = (
    (p_win_given_pam * p_pam) +
    (p_win_given_paul * p_paul) +
    (p_win_given_peter * p_peter)
)
```

```python
#Bayes: probability winner came from Pam
p_pam_given_win = (p_win_given_pam * p_pam) / p_win

print(p_pam_given_win)
```
